# Synthetic dataset

Plots the performance of different active learning strategies on a synthetic dataset.

```python
import matplotlib.pyplot as plt
import numpy as np
import pickle
import thesis

from thesis.robustsort import plots

%matplotlib inline
```

```python
thesis.setup_plotting()
```

## Load the data

```python
with open("data/synthetic-n200-r5.pickle", "rb") as f:
    data = pickle.load(f)
```

## Plot it

```python
strategies = ("simpleu", "entropy", "kldiv", "mergesort", "quicksort", "random")
n_comps = 4000
xs = np.linspace(200, n_comps, num=15, dtype=int)
```

```python
fig, ax = plt.subplots(figsize=(4.25, 3.25))

for s in strategies:
    mean, std = data[s]
    ax.errorbar(xs, mean, yerr=std, label=plots.label[s],
            linewidth=0.75, capsize=1.5,
            color=plots.c[s], linestyle=plots.ls[s],
            marker=plots.marker[s], markerfacecolor=plots.mfc[s])
#ax.set_yscale("log")
ax.legend()
ax.set_ylim(ymax=8500)
ax.set_xlabel("Number of comparisons $c$")
ax.set_ylabel("Displacement")
ax.set_title("Synthetic dataset")

fig.savefig("synthetic.pdf")
```
