# Sushi dataset

Plots the performance of different active learning strategies on the [sushi dataset](http://www.kamishima.net/sushi/).

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pickle
import thesis

from thesis.robustsort import plots
```

```python
thesis.setup_plotting()
```

## Load the data

```python
with open("data/sushi-p1.pickle", "rb") as f:
    data = pickle.load(f)
disp = data["disp"]
xs = data["xs"]
```

## Plot it

```python
fig, ax = plt.subplots(figsize=(4.25, 3.25))

for s in ("simpleu", "entropy", "kldiv", "mergesort", "quicksort", "random"):
    mean, std = disp[s]
    ax.errorbar(xs, mean, yerr=std, label=plots.label[s],
            linewidth=0.75, capsize=1.5,
            color=plots.c[s], linestyle=plots.ls[s],
            marker=plots.marker[s], markerfacecolor=plots.mfc[s])
    ax.legend()
    ax.set_ylim(ymin=0, ymax=2600)
    ax.set_xlabel('Number of comparisons $M$')
    ax.set_ylabel('Displacement')
    ax.set_title("Sushi dataset")

fig.savefig("sushi.pdf")
```
