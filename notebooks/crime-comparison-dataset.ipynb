```python
import choix
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
mat = np.transpose(
        np.loadtxt("../data/ccdata-fixed.csv", delimiter=",", dtype=int))

with open("../data/ccdata-labels.txt") as f:
    labels = list(line.strip() for line in f)
```

```python
params = choix.ilsr_pairwise_dense(mat)
```

```python
fig, ax = plt.subplots(figsize=(16, 4))
ax.set_axis_off()
ax.scatter(params, np.zeros_like(params), color="black", s=20.0)

for i, (x, label) in enumerate(sorted(zip(params, labels))):
    if i % 2 == 0:
        ax.annotate(label, (x, 0), ha="left", va="bottom", rotation=45)
    else:
        ax.annotate(label, (x, 0), ha="left", va="top", rotation=-45)

ax.axhline(0.0, color="black")
ax.scatter(max(params) + 0.5, 0, marker=">", color="black", s=100.0)
```
